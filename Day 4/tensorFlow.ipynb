```python
import numpy as np
x=np.random.random((1000,32))
y=np.random.randint(2,size=(1000,1))

dataset = ts.data.Dataset.from_tensor_slices((x,y))

dataset = dataset.shuffle(buffer_size=1000).batch(32).prefetch(ts.data.AUTOTUNE)

for batch_x,batch_y in dataset.take(1):
    print("Batch X shape:" , batch_x.shape)
    print("Batch y shape:", batch_y.shape)
```

```python
x_train = np.random.random((800,10))
y_train = np.random.randint(2,size=(800,1))
x_test = np.random.random((200,10))
y_test = np.random.randint(2,size=(200,1))
```

```python
model = ts.keras.Sequential([
    ts.keras.layers.Dense(64,activation='relu', input_shape=(10,)),
    ts.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=32,verbose=True)
```

```python
loss,accuracy = model.evaluate(x_test,y_test)
print(f"Loss: {loss:.4f}, Accuracy: {accuracy:.4f}")
```

```python
predictions=model.predict(x_test[:5])
print(predictions)
```

```python
#Required import statements

import tensorflow as tf
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
import numpy as np

#Load and preprocess data

(train_images,train_labels),(test_images,test_labels)=tf.keras.datasets.mnist.load_data()
train_images=train_images.reshape((60000,28*28)) / 255.0
test_images=test_images.reshape((10000,28*28)) / 255.0
plt.figure(figsize=(10,5))
for i in range(20):
  plt.subplot(4,5,i+1)
  plt.imshow(train_images[i].reshape(28,28),cmap='gray')
  plt.title(f"Label: {train_labels[i]}")
  plt.axis('off')
plt.suptitle('Sample MNIST Images with labels:',y=1.02)
```

```python
model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(28*28,)),
    layers.Dense(64,activation='relu'),
    layers.Dense(32,activation='relu'),
    layers.Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_images,train_labels,batch_size=30,epochs=40,validation_data=(test_images,test_labels))
```

```python
test_loss,test_acc=model.evaluate(test_images,test_labels)
test_res=test_acc*100
print(f"Test accuracy: {test_res:.2f}")
```
